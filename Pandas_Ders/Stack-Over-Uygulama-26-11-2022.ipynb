```python
# Data Filtreleme 

df = pd.read_csv('developer-survey-2019/survey_results_public.csv', index_col='Respondent')

df.head(5)
```

```python
# Uygulama-1 :

# "Maaş"ı (ConvertedComp) 70000 den yüksek olanları sırasıyla ;
# "Ülke", "Çalıştıkları Programlama Dili", "Maaş" bilgilerini getirelim.

# Maaş : 'ConvertedComp'  ,   Ülke : 'Country'  ,  
# Çalıştıkları Programlama Dili : 'LanguageWorkedWith' 
  
#    'Country' ,  'LanguageWorkedWith',  'ConvertedComp'
# Uygun filtreyi yazıp doğru sonucu getiren indexlemeyi data frame'e uygulayalım...
```

```python
my_filt = (df['ConvertedComp'] > 70000)

df.loc[my_filt, ['Country' ,  'LanguageWorkedWith',  'ConvertedComp']]
```

```python
# Uygulama-2 :

# Aşağıdaki ülkelerden birinden olanları bulan filtreyi yazalım.
# 'United States', 'India', 'United Kingdom', 'Germany', 'Canada'
# filtreyi data frame'e sadece ülke 'Country' bilgisini getirecek şekilde uygulayalım.
```

```python
country_list = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']

my_filt = df['Country'].isin(country_list)

df.loc[my_filt, 'Country' ]
```

```python
df['LanguageWorkedWith']
```

```python
filt = df['LanguageWorkedWith'].str.contains('Python', na=False)

df.loc[filt, 'LanguageWorkedWith']
```

```python
new_filt = df['LanguageWorkedWith'].str.contains('Python|HTML|Java', na=False)

df.loc[new_filt, ['LanguageWorkedWith', 'Country']]
```

```python
# Uygulama-1 :
# Ülke isimlerine göre tüm dataframe'de sorting işlemi yapmak istiyoruz.
# Bu verilerin bulunduğu sütun; 'Country'
# 1. Kalıcı sorting yapalım.
# 2. Sorting sonrası bu sütundaki ilk 50 veriyi görelim.

df.sort_values(by= 'Country', inplace=True)
df['Country'].head(150)
```

```python
# Uygulama 2: 
# Şimdi bu sorting işlemi sonrası, 'Country' ile birlikte 
# maaş verilerinin olduğu 'ConvertedComp' sütununu birlikte görüntüleyelim:
# Sadece ilk 50 satırı görelim.
# Birden fazla sütun label girerken ayrı bir [] içerisine almaya dikkat edelim...

df[['Country','ConvertedComp']].head(50)
```

```python
# Uygulama 3: 
# Ülke sıralaması alfabetik olarak sıralansın(artarak gitsin). (A,B,C....)
# Ülke sıralaması ile birlikte Maaş 'ConvertedComp' sıralaması ise en yüksekten en düşüğe doğru yani azalarak gitsin.
# Böylece herbir ülkedeki sıralama en yüksekten en düşüğe doğru olsun.
# Öncelik Ülke sıralamasında olsun.
# Sorting işlemi kalıcı olsun.
# Sorting sonrası Ülke ve Maaş sütunlarının  ilk 50 satırını görüntüleyelim.

df.sort_values(by=['Country','ConvertedComp'], ascending=[True, False], inplace=True)

df[ ['Country', 'ConvertedComp'] ].head(50)
```

```python
# Uygulama 4: 
# Maaş sütununda En yüksek 10 maaşın olduğu satırları getirmesini istersek;
# df['sütun_label'].nlargest(satır_sayısı)

df['ConvertedComp'].nlargest(10)
```

```python
# Uygulama 5:
# En yüksek 10 maaş bulunan satırları getirecek şekilde 
# tüm data frame'i görüntülemek için :
# df.nlargest( istenilen_satır_sayısı, sütun_label )

df.nlargest( 10, 'ConvertedComp' )
```

```python
df.nsmallest(10, 'ConvertedComp')
```
