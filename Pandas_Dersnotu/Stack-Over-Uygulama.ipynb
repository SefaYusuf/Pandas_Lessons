### mean - median - mode

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').sum())
```

```python
# concat()

python_df = pd.concat([country_resp,country_uses_python], axis = 'columns', sort = False)

python_df
```

```python
python_df.rename(columns = {'Country' : 'NumberRespond','LanguageWorkedWith' : 'NumKnowPython'}, inplace = True)
```

```python
python_df
```

```python
# Şimdi bu dataframe'e herbir ülkeden Python bilenlerin yüzde değerlerini
# ayrı bir sütun olarak ekleyelim :

python_df['PctKnowsPython'] = python_df['NumKnowPython']/python_df['NumberRespond']*100
```
